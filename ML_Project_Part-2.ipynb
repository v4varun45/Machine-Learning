```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib .pyplot as plt
```

```python
%matplotlib inline
```

```python
# Loading the dataset

Csv_dataset=pd.read_csv("petrol_consumption.csv")
```

```python
Csv_dataset.isnull()
```

```python
Csv_dataset.head()
```

```python
X = Csv_dataset.drop('Petrol_Consumption', axis=1)
y = Csv_dataset['Petrol_Consumption']
```

```python
X.head()
```

```python
y.head()
```

```python
Csv_dataset.describe()
```

```python
# Histogram
Csv_dataset.hist()
plt.show()
```

```python
# Density plot
Csv_dataset.plot(kind='density',subplots=True,layout=(3,3),sharex=False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.50,random_state=0)
```

```python
# Training the model

from sklearn.tree import DecisionTreeRegressor
Reg = DecisionTreeRegressor()
Reg.fit(X_train,y_train)
y_pred = Reg.predict(X_test)
y_pred
```

```python
# Creating a dataframe

Data_frame = pd.DataFrame({"Actual":y_test, "predicted":y_pred})
Data_frame
```

```python
plt.figure(figsize=(8, 10))
ax = sns.distplot(y_test, hist=False, color="r", label="Actual Values")
sns.distplot(y_pred, hist=False, color="g", label="Predicted Values" , ax=ax)

plt.title('Actual vs Predicted Values')
plt.legend(loc="upper right")

plt.show()
plt.close()
```

```python
# Finding the Errors

from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Square Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# 2) Random Forest Regressor

```python
# Loading the dataset

dataset_Forest=pd.read_csv("petrol_consumption.csv")
```

```python
dataset_Forest.head()
```

```python
X = dataset_Forest.drop('Petrol_Consumption', axis=1)
y = dataset_Forest['Petrol_Consumption']
```

```python
X.head()
```

```python
y.head()
```

```python
dataset_Forest.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.50,random_state=0)
```

```python
# Training the model

from sklearn.ensemble import RandomForestRegressor
For_Reg = RandomForestRegressor(n_estimators=50, random_state=0)
For_Reg.fit(X_train,y_train)
y_pred = For_Reg.predict(X_test)
y_pred
```

```python
# Creating a dataframe

Data_frame01 = pd.DataFrame({"Actual":y_test, "predicted":y_pred})
Data_frame01
```

```python
plt.figure(figsize=(8, 10))
ax = sns.distplot(y_test, hist=False, color="r", label="Actual Values")
sns.distplot(y_pred, hist=False, color="g", label="Predicted Values" , ax=ax)

plt.title('Actual vs Predicted Values')
plt.legend(loc="upper right")

plt.show()
plt.close()
```

```python
# Finding the Errors

from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Square Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
